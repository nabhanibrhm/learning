**Extract Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\Purwadhika\belajar aja\learning\capstone_2\data\Airbnb Listings Bangkok.csv')
df.drop(['Unnamed: 0'], axis=1 ,inplace=True)
df.head()
```

**Data Dictionary**

* id: Identitas unik Airbnb untuk tempat tersebut
* name: Nama daftar
* host_id:Identitas unik Airbnb untuk tuan rumah/pengguna
* neighborhood:Lingkungan yang diberi kode geografis menggunakan garis lintang dan bujur terhadap lingkungan yang ditentukan oleh shapefile digital terbuka atau publik
* latitude: Proyeksi Sistem Geodesi Dunia (WGS84) untuk garis lintang dan garis bujur.
* longitude:  Proyeksi Sistem Geodesi Dunia (WGS84) untuk garis lintang dan garis bujur.
* room_type:
* price: Harga harian dalam mata uang lokal. Catatan, tanda $ dapat digunakan terlepas dari lokasinya
* minimum_nights: jumlah minimum masa inap malam untuk tempat tersebut
* number_of_reviews: Jumlah ulasan yang dimiliki listingan
* last_review: Tanggal review terakhir/terbaru
* reviews_per_month: jumlah review per bulan
* calculated_host_listings_count:
* availability_365: berapa hari dalam setahun kamar/ rumah dapat dipakai
* number_of_reviews_ltm: number of reviews that the property received in the last 12 months

```python
room_type_counts = df['room_type'].value_counts()

plt.bar(room_type_counts.index, room_type_counts.values, color='skyblue')

for i, value in enumerate(room_type_counts.values):
    plt.text(i, value + 0.1, str(value), ha='center', va='bottom')

plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Bar Chart of Room Type')
plt.show()
```

```python
#top 10 hosting

top_host=df.host_id.value_counts().head(10)
```

```python
#the most popular neighborhood for Airbnb listings in Bangkok
top_neighbourhood=df.neighbourhood.value_counts().head(10)
top_neighbourhood
```

```python
top_neighbourhood=df['neighbourhood'].value_counts().head(5)

plt.bar(top_neighbourhood.index, top_neighbourhood.values, color='green')

for i, value in enumerate(top_neighbourhood.values):
    plt.text(i, value + 0.1, str(value), ha='center', va='bottom')

plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Bar Chart of Top 5 Nieghborhood')
plt.show()
```
