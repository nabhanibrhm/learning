**Extract Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\Purwadhika\belajar aja\learning\capstone_2\data\Airbnb Listings Bangkok.csv')
df.drop(['Unnamed: 0'], axis=1 ,inplace=True)
df.head()
```

**Data Dictionary**

* id: Identitas unik Airbnb untuk tempat tersebut
* name: Deskripsi porperti
* host_id: Identitas unik Airbnb untuk tuan rumah/pengguna
* host_name: Orang yang menyajikan properti
* neighborhood:Lingkungan sekitar wilayah
* latitude: Proyeksi Sistem Geodesi Dunia (WGS84) untuk garis lintang dan garis bujur.
* longitude:  Proyeksi Sistem Geodesi Dunia (WGS84) untuk garis lintang dan garis bujur.
* room_type: Jenis porperti
* price: Harga sewa per-malam
* minimum_nights: jumlah minimum masa inap malam untuk tempat tersebut
* number_of_reviews: Jumlah ulasan yang dimiliki listingan
* last_review: Tanggal review terakhir/terbaru
* reviews_per_month: Jumlah review per bulan
* calculated_host_listings_count: Jumlah tamu
* availability_365: berapa hari dalam setahun kamar/ rumah dapat dipakai
* number_of_reviews_ltm: Jumlah ulasan yang diterima properti dalam 12 bulan terakhir

Penjelasan
* *missing value* di kolom `name`, `host_name` memiliki presentase yang rendah (<10%), dengan missing value di kolom `last_review` dan `reviews_per_month` mencapai 36.5%

```python
#replace NaN menjadi 0
df['last_review'].fillna(0, inplace=True)
df['reviews_per_month'].fillna(0, inplace=True)
```

* Karna *missing value* di kolom `last_review` dan `reviews_per_month` mencapai 36.5% maka akan digantikan valuenya menjadi 0 karena memang belum ada user yang me-review tempat tersebut dan jika data di drop tidak efektif karena mengurangi setenah dari jumlah data sebenarnya

```python
room_type_counts = df['room_type'].value_counts()

plt.bar(room_type_counts.index, room_type_counts.values, color='skyblue')

for i, value in enumerate(room_type_counts.values):
    plt.text(i, value + 0.1, str(value), ha='center', va='bottom')

plt.xlabel('Room Type')

plt.ylabel('Count')
plt.title('Bar Chart of Room Type')
plt.show()
```
