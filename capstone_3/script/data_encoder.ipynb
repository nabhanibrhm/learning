```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.read_csv(r'D:\Gawe\projects_learning\capstone_3\data\data_travel_insurance_prep.csv',sep='|')
df.head()
```

```python
category_columns = df[['Agency', 'Agency Type', 'Distribution Channel', 'Product Name', 'Claim', 'Destination']]
```

```python
label_encoder = LabelEncoder()

category_transform = category_columns.apply(LabelEncoder().fit_transform)

category_transform
```

```python
df.drop(['Agency', 'Agency Type', 'Distribution Channel', 'Product Name', 'Claim', 'Destination'], axis = 1, inplace = True)
df = pd.concat([df, category_transform], axis=1)
df
```

```python
df.to_csv(r'D:\Gawe\projects_learning\capstone_3\data\data_travel_insurance_transform.csv',sep='|', index=False)
```
